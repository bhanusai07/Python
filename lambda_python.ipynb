```python
sai = lambda x,y:x+y
print(sai(2,3))
```

```python
a = [10,54,85,55]
b= list(map(lambda x:x+10,a))
print(b)
```

```python
ticket_without_hike=[250,150,300]
with_hike=(map(lambda x:x+10,ticket_without_hike))
s= list(with_hike)
print("movie with hike: ",ticket_without_hike)
print("movie with hike: ",s)
```

```python
from functools import reduce
nums=[85,52,68,64]
a=(reduce(lambda x,y:x+y,nums))
print(a)
```
