```python
balance = 5400
pin = 1234
entered_pin=int(input("Enter the pin:"))
if entered_pin ==pin:
  print("1.Balance")
  print("2.deposit")
  print("3.Withdraw")
  choice = int(input("Enter your choice:"))
  if choice == 1:
    print("Available Balance:",balance)
  elif choice == 2:
    amount= int(input("enter the amount:"))
    balance = balance+amount
    print("amount deposited")
    print("Available balance:",balance)
  elif choice == 3:
    amount = int(input("enter the amount:"))
    balance = balance-amount
    print("available balance:",balance)
  else:
    print("Invalid choice")
else:
  print("Invalid pin")
```
